```python
# end_to_end_flow.csv is a csv extracted usigng live data capture, this is 1 flow of the data, the onboardig webpage has 5 different sections each having different branching section(YES?NO(2^n possibilities)), different repeatable sections (Benificiary, jooint partner,, isbeneficiary, permananentaddresssameasmailing and many more (n! factorial * m possibilities)), and various dropdowns for county, code, occupations, salary range etc. This poses a problem in creating multiple test cases (probablities could exceed millions). A simple example is all_elements.csv(access and have a look) where i have tried to select all elements in the all sections atleast one time for eleemtnname+Action capture, the rult is 745 rows with diffrent eleemtn data+actions+xpaths (even here a few thifns have been left unselected). So it is pretty clear that making different probabilities on all_elements.csv is impossible.

# The suggested solution:
# We provide an LLM model, the all_element.csv as context, the model will learn the grouping (Groups columsn to identify different pages), heirarchy of data flow and subsections+options, semantic relationship of heirarchial flow, branches, repeatable sections and dropdown options. Then we use Advanced prompt engineering and ask the model to generate us a new testcase csv.
# Example. We ask model to create a csv of test case having Gender as male, including 2 benificiary, 3 joint acctount etc etc etc, although a little difficult, this is the best way for a QA to generate different test case scenarios. This will give us multiple csv as needed by the user with different scenarios, sub section , repetiton setions and flow included. All controlable through prompt and not probability.
```
