### Faker

Objective is to generate a fake data everytime we call a label

```python
from faker import Faker
import random
```

```python
fake = Faker()
```

```python
#field type detector, this tells faker that if dict contains a label = "name", then it is a name field, if "email", then it is email field 
FIELD_PATTERNS = {
    'name': ['name', 'firstname', 'lastname', 'middlename'],
    'email': ['email', 'mail'],
    'phone': ['phone', 'mobile', 'tel', 'mobilenumber'],
}

def detect_field_type(label):
    label_lower = label.lower().replace('_', '').replace('-', '')
    for field_type, patterns in FIELD_PATTERNS.items():
        for pattern in patterns:
            if pattern in label_lower:
                return field_type
    return 'text'
```

```python
# value generator
def generate_values(field_type, original_value, count=10):
    if field_type == 'name':
        values = [original_value, '', fake.first_name().upper(), '123', '@#$%', 'A'*51, 'A', ' ']
    elif field_type == 'email':
        values = [original_value, '', fake.email(), 'invalid', 'test@', '@gmail.com', 'a@b.c']
    elif field_type == 'phone':
        values = [original_value, '', fake.numerify('###-####'), 'abc', '1', '123-4567-8901']
    else:
        values = [original_value, '', fake.word().upper(), '123', '@#$%', 'A'*50]
    
    while len(values) < count:
        values.append(fake.first_name().upper() if field_type == 'name' else fake.word())
    return values[:count]
```

```python
print(generate_values('name', 'PRAKYATH'))
```

```python
def generate_test_cases(df, num_rows=100):
    #clean data - fill NaN clicks with clicks
    df_clean = df.copy()
    df_clean['Value'] = df_clean.apply(
        lambda row: 'Click' if row['Action'] == 'click' else row['Value'],
        axis=1
    )

    elements = df_clean['Label'].tolist()
    actions = df_clean['Action'].tolist()

    # Find which column have 'change' action (need value generation)
    change_indices = [i for i, action in enumerate(actions) if action == 'change']

    # generate value pools for each change field 
    value_pools = {}
    for idx in change_indices:
        label = elements[idx]
        original_value = df_clean.iloc[idx]['Value']
        field_type = detect_field_type(label)
        value_pools[idx] = generate_values(field_type, original_value, num_rows)
        print(f"{label}: detected as '{field_type}'")

    # generate test case
    test_rows = []
    for row_num in range(num_rows):
        row = []
        for i in range(len(elements)):
            if i in change_indices:
                row.append(value_pools[i][row_num % len(value_pools[i])])
            else:
                row.append('Click')
        test_rows.append(row)


    # Build final DataFrame
    result_data = [elements, actions] + test_rows
    result_df = pd.DataFrame(result_data)
    result_df.index = ["Element", "Action"] + [f'TC_{str(i+1).zfill(3)}' for i in range(num_rows)]

    return result_df
    

```

```python
result = generate_test_cases(df, num_rows=100)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
result.head(105)
```
