```python
import pandas as pd
```

```python
df = pd.read_csv("xpaths_20260127_175205.csv")
```

```python
df = df[["Label", "Action", "Value", "XPath"]]
```

```python
df
```

```python
# Step 1: Clean data
df_clean = df[['Label', 'Action', 'Value', "XPath"]].copy()

df_clean['Action'] = df_clean['Action'].replace('change', 'input')


# Step 2: Replace NaN with 'Click' for click actions
df_clean['Value'] = df_clean.apply(
    lambda row: 'Click' if row['Action'] == 'click' else row['Value'],
    axis=1
)


# Step 3: Build the three rows
labels = df_clean['Label'].tolist()
actions = df_clean['Action'].tolist()
values = df_clean['Value'].tolist()
xpath = df_clean["XPath"].to_list()

# Step 4: Create result dataframe
rows = [
    ['Description'] + labels,           # Row 0: Header
    ['Action'] + actions,               # Row 1: Actions
    ['TestCase_001'] + values,          # Row 2: Values
    ['XPath'] + xpath
]

result = pd.DataFrame(rows)
# Show all columns
pd.set_option('display.max_columns', None)

# Optional: also expand column width
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)









result
```

```python
result.to_csv("end_to_end_flow_output.csv")
```
